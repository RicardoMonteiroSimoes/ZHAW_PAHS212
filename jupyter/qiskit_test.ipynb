```python
# imports
# if it fails here, you might not have installed all the required packages
import numpy as np
from qiskit import QuantumCircuit, transpile
from qiskit.visualization import plot_histogram
from qiskit.providers.aer import QasmSimulator
```

Let's construct a circuit containing 2 qubits and put them in a super position

```python
circuit = QuantumCircuit(2,2)
circuit.h(0)
circuit.h(1)
circuit.measure([0,1],[0,1])
circuit.draw('mpl')
```

Let's now build the simulator, compule the circuit and run it. The result of this circuit should be a probability of ~1/4 for each of the following states: _{00,01,10,11}_

```python
simulator = QasmSimulator()
compiled_circuit = transpile(circuit, simulator)
job = simulator.run(compiled_circuit, shots=3000)
result = job.result()
counts = result.get_counts(circuit)
plot_histogram(counts)
```
