```python
import numpy as np
import matplotlib.pyplot as plt

#create folder for figures
import os
path_boxplots = './plots_decision_boundaries/boxplots/'
try:
    os.makedirs(path_boxplots)
except:
    print('path already exists!')

# single qubit circuits 
def plot_boxplot_single(scores, title, iterations, path, save=False):
    tests = []
    run_count = len(scores[0])
    for classifier_score in scores:
        tests.append(classifier_score)
    min = np.amin(tests)-0.01
    max = np.amax(tests)+0.01
    fig = plt.figure()
    gs = fig.add_gridspec(1, 1, hspace=0, wspace=0)
    (ax1) = gs.subplots(sharex='col', sharey='row')
    ax1.set_ylim([min,max])
    fig.suptitle(title)
    ax1.set_title(str(run_count) + " runs / " + str(iterations) + " iterations per run")
    ax1.set_xlabel('Classifier')
    ax1.set_ylabel('Accuracy')
    ax1.boxplot(tests, showfliers=True, sym='k.')
    plt.xticks([1, 2, 3], ['MLP', 'QC1', 'QC2'])
    ax1.yaxis.grid(True, linestyle='-', which='major', color='lightgrey',
               alpha=0.5)
    if(save):
        plt.savefig(path+ '1qubit_' + title.replace(" ", "_")+ '_' + str(run_count) + 'runs_' + str(iterations) + '.png', format="png", transparent=True)
    plt.show()
    plt.close()
    
def create_box_plots_single(scores, iterations):
    linear_scores = []
    circle_scores = []
    moon_scores = []
    for classifier in scores:
        linear_scores.append(classifier[:,0])
        circle_scores.append(classifier[:,1])
        moon_scores.append(classifier[:,2])
    
    plot_boxplot_single(linear_scores, "Linearly separable data", iterations, path_boxplots, True)
    plot_boxplot_single(circle_scores, "Circle shaped data", iterations, path_boxplots, True)
    plot_boxplot_single(moon_scores, "Moon shaped data", iterations, path_boxplots, True)

# 2 qubit circuits 
def plot_boxplot_2qubits(scores, title, iterations, path, save=False):
    tests = []
    run_count = len(scores[0])
    for classifier_score in scores:
        tests.append(classifier_score)
    min = np.amin(tests)-0.01
    max = np.amax(tests)+0.01
    fig = plt.figure()
    gs = fig.add_gridspec(1, 1, hspace=0, wspace=0)
    (ax1) = gs.subplots(sharex='col', sharey='row')
    ax1.set_ylim([min,max])
    fig.suptitle(title)
    ax1.set_title(str(run_count) + " runs / " + str(iterations) + " iterations per run")
    ax1.set_xlabel('Classifier')
    ax1.set_ylabel('Accuracy')
    ax1.boxplot(tests, showfliers=True, sym='k.')
    plt.xticks([1, 2, 3, 4], ['MLP', 'QC1', 'QC2', 'QC3'])
    ax1.yaxis.grid(True, linestyle='-', which='major', color='lightgrey',
               alpha=0.5)
    if(save):
        plt.savefig(path+'2qubits_' + title.replace(" ", "_")+ '_' + str(run_count) + 'runs_' + str(iterations) + '.png', format="png", transparent=True)
    plt.show()
    plt.close()

def create_box_plots_2qubits(scores, iterations):
    linear_scores = []
    circle_scores = []
    moon_scores = []
    for classifier in scores:
        linear_scores.append(classifier[:,0])
        circle_scores.append(classifier[:,1])
        moon_scores.append(classifier[:,2])
    
    plot_boxplot_2qubits(linear_scores, "Linearly separable data", iterations, path_boxplots, True)
    plot_boxplot_2qubits(circle_scores, "Circle shaped data", iterations, path_boxplots, True)
    plot_boxplot_2qubits(moon_scores, "Moon shaped data", iterations, path_boxplots, True)

# 2 qubit circuits 3 layers
def plot_boxplot_2qubits_3layers(scores, title, iterations, path, save=False):
    tests = []
    run_count = len(scores[0])
    for classifier_score in scores:
        tests.append(classifier_score)
    min = np.amin(tests)-0.01
    max = np.amax(tests)+0.01
    fig = plt.figure()
    gs = fig.add_gridspec(1, 1, hspace=0, wspace=0)
    (ax1) = gs.subplots(sharex='col', sharey='row')
    ax1.set_ylim([min,max])
    fig.suptitle(title)
    ax1.set_title(str(run_count) + " runs / " + str(iterations) + " iterations per run")
    ax1.set_xlabel('Classifier')
    ax1.set_ylabel('Accuracy')
    ax1.boxplot(tests, showfliers=True, sym='k.')
    plt.xticks([1, 2, 3, 4], ['MLP', 'QC1', 'QC2', 'QC3'])
    ax1.yaxis.grid(True, linestyle='-', which='major', color='lightgrey',
               alpha=0.5)
    if(save):
        plt.savefig(path+'2qubits_3layers_' + title.replace(" ", "_")+ '_' + str(run_count) + 'runs_' + str(iterations) + '.png', format="png", transparent=True)
    plt.show()
    plt.close()

def create_box_plots_2qubits_3layers(scores, iterations):
    linear_scores = []
    circle_scores = []
    moon_scores = []
    for classifier in scores:
        linear_scores.append(classifier[:,0])
        circle_scores.append(classifier[:,1])
        moon_scores.append(classifier[:,2])
    
    plot_boxplot_2qubits_3layers(linear_scores, "Linearly separable data", iterations, path_boxplots, True)
    plot_boxplot_2qubits_3layers(circle_scores, "Circle shaped data", iterations, path_boxplots, True)
    plot_boxplot_2qubits_3layers(moon_scores, "Moon shaped data", iterations, path_boxplots, True)
```
