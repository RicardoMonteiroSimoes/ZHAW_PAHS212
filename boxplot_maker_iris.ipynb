```python
def plot_boxplots(path, run, title, circuit_index, save=False):
    min = np.amin(run)-0.01
    max = np.amax(run)+0.01
    fig = plt.figure()
    gs = fig.add_gridspec(1, 2, hspace=0, wspace=0)
    (ax1, ax2) = gs.subplots(sharex='col', sharey='row')
    ax1.set_ylim([min,max])
    ax2.set_ylim([min,max])
    fig.suptitle(title)
    ax1.set_title('Training')
    ax1.boxplot(run[0], showfliers=True, sym='k.')
    ax2.set_title('Testing')
    ax2.boxplot(run[1], showfliers=True, sym='k.')
    if(save):
        plt.savefig(path+title.replace(" ", "_")+'.svg', format="svg", transparent=True)
    plt.close()
```

```python
def comparison_all_circuits(measurments,title, path, save=True):
    tests = []
    for run in measurments:
        tests.append(run[1])
    min = np.amin(tests)-0.01
    max = np.amax(tests)+0.01
    fig = plt.figure()
    gs = fig.add_gridspec(1, 1, hspace=0, wspace=0)
    (ax1) = gs.subplots(sharex='col', sharey='row')
    ax1.set_ylim([min,max])
    fig.suptitle(title)
    ax1.set_title('Testing')
    ax1.boxplot(tests, showfliers=True, sym='k.')
    if(save):
        plt.savefig(path+title.replace(" ", "_")+'.svg', format="svg", transparent=True)
    plt.close()
```
